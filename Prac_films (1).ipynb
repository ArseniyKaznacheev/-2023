# 1. Построение ядерных оценок плотности

Ядерная оценка плотности (Kernel Density Estimation, KDE) — это непараметрический способ оценки плотности случайной величины. Ядерная оценка плотности является задачей сглаживания данных, когда делается заключение о совокупности, основываясь на конечных выборках данных.

Данный метод реализован в библиотечной функции displot.

1. Сначала для каждой точки наблюдений строится график нормального распределения с центром в этой точке.
2. Значения каждой нормальной кривой суммируются, а затем полученное значение нормализуется так, чтобы площадь под результирующей кривой стала равна 1.
3. Полученный график сглаживается в зависимости от выбора полосы пропускания.

# 3. Проверка на выбросы с помощью критерия Граббса и Q-теста Диксона

Тесты **Граббса** и **Диксона** используется для определения наличия выбросов в наборе данных.

Выбросы в данных — это значения, которые сильно отличаются от большинства других значений в наборе данных. Они могут возникнуть из-за ошибок ввода, манипуляций с данными или аномальных процессов. Выбросы могут вносить существенные искажения в результаты анализа, поэтому важно определить и обработать их перед анализом данных.

##Тест Граббса

При про­верке на выброс $X_{n}$ стати­стика критерия Граббса имеет вид $G_{n} = \frac{(X_{n} - \bar X)}{S}$, где \\
$\bar X = \frac{1}{n} \sum\limits_{j=1}^{n}X_{j};$ \\
$S^{2}=\frac {1}{n-1} \sum\limits_{j=1}^{n}(X_{j}-X)^{2}$

```python
boxplot(data[which(data$Subject == "Drama"),]$Length, horizontal = TRUE, xlab="Продолжительность", main = "Drama")
```

##Тест Диксона

 Чтобы применить Q-тест для неверных данных, надо расположить данные в порядке возрастания значений и вычислить Q, как определено: \\
 $Q = \frac{gap}{range}$, где \\
$gap$ - модуль разности между предполагаемым выбросом и ближайшим к нему другим объектом. \\
$range$ - разность максимального и минимального значения в выборке.

## Тест $\chi ^2$

```python
# Python
from scipy.stats import chisquare

obs = data['Length']
exp = [data['Length'].mean()]*data.shape[0]
print(chisquare(f_obs=obs, f_exp=exp))

obs = data['Popularity']
exp = [data['Popularity'].mean()]*data.shape[0]
print(chisquare(f_obs=obs, f_exp=exp))

obs = data['Year']
exp = [data['Year'].mean()]*data.shape[0]
print(chisquare(f_obs=obs, f_exp=exp))
```

##Точный тест Фишера:

Проверим однородность данных внутри некоторых жанров.

```python
# python
from scipy.stats import fisher_exact
tmp1 = data[(data.Subject == "Drama")|
            (data.Subject == "Action")].Awards.value_counts()

tmp2 = data[(data.Subject == "Comedy")|
            (data.Subject == "Horror")|
            (data.Subject == "Mystery")].Awards.value_counts()

m = [[tmp1['Yes'], tmp1['No']],
     [tmp2['Yes'], tmp2['No']]]

print(fisher_exact(m))
```

```python
# R
tmp1 <- table(data[data$Subject == "Drama" |
                   data$Subject == "Action",]$Awards)
tmp2 <- table(data[data$Subject == "Comedy" |
                   data$Subject == "Horror" |
                   data$Subject == "Mystery", ]$Awards)

m <- matrix(c(tmp1['Yes'], tmp1['No'], tmp2['Yes'], tmp2['No']),
            nrow=2, ncol=2, byrow=TRUE)
print(fisher.test(m))
```

##Тест МакНемара:

```python
# Python
from statsmodels.stats.contingency_tables import mcnemar
m1 = [[sum(tmp1) - tmp1['No'], tmp1['No']],
      [sum(tmp2) - tmp2['No'], tmp2['No']]]

print(mcnemar(m1, exact=False))
```

```python
# R
m1 <- matrix(c(sum(tmp1) - tmp1['No'], tmp1['No'], sum(tmp2) - tmp2['No'], tmp2['No']),
            nrow=2, ncol=2, byrow=TRUE)
print(mcnemar.test(m1))
```

#12. Подогнать регрессионные модели (в том числе, нелинейные) к данным, а также оценить качество подобной аппроксимации.

Линейная модель с $L^{1}$ регуляризацией.

```python
library(glmnet)
```

```python
# R

index <- c('Subject', 'Length', 'Year', 'Popularity', 'Awards', 'Budget',
           'revenue')

sample <- sample(c(TRUE, FALSE), 1592, replace=TRUE, prob=c(0.7,0.3))
train  <- data[sample, ]
test   <- data[!sample, ]

cv_model <- cv.glmnet(data.matrix(train[,index]),
                      data.matrix(train$Length),
                      alpha = 1)

best_lambda <- cv_model$lambda.min

plot(cv_model)
```

```python
# R
best_model <- glmnet(data.matrix(train[,index]),
                     data.matrix(train$Length),
                     alpha=1,
                     lambda=best_lambda)

pred <- predict(best_model, s=best_lambda, newx=data.matrix(test[,index]))

print(paste("test RMSE = ", sqrt(mean((pred - test$Length)**2))))
```
